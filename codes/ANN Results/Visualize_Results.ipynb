# Visualizing ANN Training Results

```python
import numpy as np
import pandas as pd
import os
os.chdir('C:\\Users\\rober\\OneDrive\\Documents\\STAT 6500\\Project\\NewsArticleClassification\\codes\\ANN Results')  ## Set working directory
                                                                                                                      ## to be 'ANN Results'
import matplotlib.pyplot as plt
```

## Learning Rate $ = 0.01$

```python
## Get the reduced dataset training histories
history_bpmll_FF_lr01_df = pd.read_json('Reduced Data Eval Metrics/BPMLL Feed Forward/history_bpmll_FF_lr01.json')
history_bpmll_RNN_lr01_df = pd.read_json('Reduced Data Eval Metrics/BPMLL RNN/history_bpmll_RNN_lr01.json')
history_ce_FF_lr01_df = pd.read_json('Reduced Data Eval Metrics/Cross Entropy Feed Forward/history_ce_FF_lr01.json')
history_ce_RNN_lr01_df = pd.read_json('Reduced Data Eval Metrics/Cross Entropy RNN/history_ce_RNN_lr01.json')
```

```python
## Visualize the 'val_hamming_loss' histories for each model in a single plot
plt.plot(history_bpmll_FF_lr01_df['val_hamming_loss'], label = 'BPMLL FF')
plt.plot(history_bpmll_RNN_lr01_df['val_hamming_loss'], label = 'BPMLL RNN')
plt.plot(history_ce_FF_lr01_df['val_hamming_loss'], label = 'CE FF')
plt.plot(history_ce_RNN_lr01_df['val_hamming_loss'], label = 'CE RNN')
plt.xlabel('Epoch', fontsize = 12)
plt.ylabel('Validation Hamming Loss', fontsize = 12)
plt.legend()
plt.title('Hamming Loss vs Epochs (Reduced Dataset)', fontsize = 15)
```

```python
## Get and print the validation hamming loss dataframe using learned threshold functions
val_hamming_loss_withThreshold_lr01_df = pd.read_json('Reduced Data Eval Metrics/val_hamming_loss_withThreshold_lr01.json')
val_hamming_loss_withThreshold_lr01_df
```

```python
## Get the full dataset training histories
history_ce_FF_lr01_full_df = pd.read_json('Full Data Eval Metrics/Cross Entropy Feed Forward/history_ce_FF_lr01_full.json')
history_ce_RNN_lr01_full_df = pd.read_json('Full Data Eval Metrics/Cross Entropy RNN/history_ce_RNN_lr01_full.json')
```

```python
## Visualize the 'val_hamming_loss' histories for each model in a single plot
plt.plot(history_ce_FF_lr01_full_df['val_hamming_loss'], label = 'CE FF')
plt.plot(history_ce_RNN_lr01_full_df['val_hamming_loss'], label = 'CE RNN')
plt.xlabel('Epoch', fontsize = 12)
plt.ylabel('Validation Hamming Loss', fontsize = 12)
plt.legend()
plt.title('Hamming Loss vs Epochs (Reduced Dataset)', fontsize = 15)
```
